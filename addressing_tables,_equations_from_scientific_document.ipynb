## Extraction of Tables from picture

image.png

```python
llm = ChatGoogleGenerativeAI(model="gemini-pro-vision")

# PNG

image = PIL.Image.open('/content/20-Table1-1.png')

hmessage1 = HumanMessage(
    content=[
        {
            "type": "text",
            "text": "In the picture you will see observation tables from research work from scientific document. I want u to send me the table in structured json format",
        },  # You can optionally provide text parts
        {"type": "image_url", "image_url": image},
    ]
)
message1 = llm.invoke([hmessage1])
print(message1.content)
```

## Grabbing photos from internet && Asking questions on tables

```python
import requests
from IPython.display import Image

image_url = "https://www.pomona.edu/sites/default/files/bio40-lab-2.jpg"
content = requests.get(image_url).content
Image(content)
```

```python
from langchain_core.messages import HumanMessage
from langchain_google_genai import ChatGoogleGenerativeAI

llm = ChatGoogleGenerativeAI(model="gemini-pro-vision")
# example
message = HumanMessage(
    content=[
        {
            "type": "text",
            "text": "What's this image about?",
        },  # You can optionally provide text parts
        {"type": "image_url", "image_url": image_url},
    ]
)
llm.invoke([message])
```
